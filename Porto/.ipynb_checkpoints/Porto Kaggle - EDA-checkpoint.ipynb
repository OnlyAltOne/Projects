#### Check for null values

The competition website has specified that all nulls have been filled with -1 values. We can replace them to see the true distribution. 

There are a few options to adress these. 
    - Leave them alone
    - Impute with mean/median/knn function of most similar neighbors
    - Drop samples/columns which contain a certain % of nulls
    
Though there are rules of thumb like filling a feature with more than 30-40% null values is unadvised, the best approach is evaluated by the affect it has on the model performance down the line.

We'll need to address these features for sure (above 5% null):
    - ps_car_03_cat
    - ps_car_05_cat
    - ps_reg_03
    - ps_car_14

```python
train_na = train.replace(-1,np.NaN)
```

```python
null_cols = []
for col in train_na.columns:
    if train_na[col].isnull().sum() > 0:
        print('feature: {}\t\t{}\tnulls\t\t{:.3f}% of data'.format(col,train_na[col].isnull().sum(),train_na[col].isnull().sum()/train.shape[0]))
    if train_na[col].isnull().sum()/train_na.shape[0] > .05:
        null_cols.append(col)
```

#### Data Dictionary or Information Provided

There isn't much which Porto Seguro has provided upfront. They've mentioned the feature prefixes relate to the type of information and the suffixes to the type of data within the column. 

Prefixes:

    ind = customer related information
    reg = region related information
    car = car related information
    calc = calculations

Suffixes:

    bin = binary
    cat = categorical
    unspecified = continuous or ordinal

```python
cat_cols = []
bin_cols = []
ord_cont_cols = []
for col in train.columns:
    if 'cat' in col:
        cat_cols.append(col)
    elif 'bin' in col:
        bin_cols.append(col)
    elif 'target' in col or 'id' in col:
        pass
    else:
        ord_cont_cols.append(col)
```

```python
cont_cols = []
ord_cols = []

for col in ord_cont_cols:
    if train[col].dtype == 'int64':
        train['{}_ord'.format(col)] = train[col]
        train.drop(col,axis=1,inplace=True)
        ord_cols.append('{}_ord'.format(col))
    else:
        cont_cols.append(col)
```

#### Correlation with target

A good start to investigating features and their relationships is to look at their correlations. Lets start with the target variable in mind.

By playing around with the top x number of features you can start feeling which features are more influencial and which less. 

    calc features seem to be least useful
    reg and car more so than ind


```python
target_corr = np.abs(train.corr()['target']).sort_values(ascending=False)
```

```python
x = 5
def top_corrs(x):
    top_corr_feats = [feat for feat in target_corr.iloc[1:x+1,].index]
    car, ind, reg, calc = 0, 0, 0, 0
    top_car_cols, top_reg_cols, top_ind_cols, top_calc_cols = [], [], [], []
    for feat in top_corr_feats:
        if 'car' in feat:
            car += 1
            top_car_cols.append(feat)
        if 'ind' in feat:
            ind += 1
            top_ind_cols.append(feat)
        if 'reg' in feat:
            reg += 1
            top_reg_cols.append(feat)
        if 'calc' in feat:
            calc += 1
            top_calc_cols.append(feat)
    
    print('--- Top {} correlations with target ---'.format(x))
    print('\t{} of {} car features'.format(car,len(car_cols)))
    print('\t{} of {} ind features'.format(ind,len(ind_cols)))
    print('\t{} of {} reg features'.format(reg,len(reg_cols)))
    print('\t{} of {} calc features'.format(calc,len(calc_cols)))
    
    return [top_car_cols, top_reg_cols, top_ind_cols, top_calc_cols]
#     print('{} of {} total car features are in the top {} highest correlated features'.format(car,len(car_cols),x))
#     print('{} of {} total ind features are in the top {} highest correlated features'.format(ind,len(ind_cols),x))
#     print('{} of {} total reg features are in the top {} highest correlated features'.format(reg,len(reg_cols),x))
#     print('{} of {} total calc features are in the top {} highest correlated features'.format(calc,len(calc_cols),x))
```

```python
for x in range(5,20,5):
    top_corrs(x)
```

```python
def highlight_corr(data,top_cols):
    corr = data.corr()
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True
    plt.figure(figsize=(16,12))
    plt.title('Pearson correlation of features',size=20)
    ax = sns.heatmap(corr, mask=mask, linewidths=0.1, vmax=1.0, 
                square=True, cmap='YlGnBu', linecolor='white', 
                annot=True, cbar_kws={'shrink': .5})

    for label in ax.get_yticklabels():
        if label.get_text() in top_cols:
            label.set_size(12)
            label.set_weight("bold")
            label.set_color("red")

    for label in ax.get_xticklabels():
        if label.get_text() in top_cols:
            label.set_size(12)
            label.set_weight("bold")
            label.set_color("red")
```

```python
from scipy.stats import pearsonr
import pandas as pd

def calculate_pvalues(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(pearsonr(df[r], df[c])[1], 4)
    return pvalues
```

```python
def create_onehot(train):
    onehot_cols = []
    for col in [col for col in train.columns[train.columns.str.contains('cat')]]:
        onehot = pd.get_dummies(train[col],prefix=col,dummy_na=np.NaN)
        for new_col in onehot.columns:
            onehot_cols.append(new_col)
        train.drop(col,axis=1,inplace=True)
        train = pd.concat([train,onehot],axis=1)
    return train
```

```python
def transform_df(train):
    train_na = train.replace(-1,np.NaN)
    
    calc_cols = [col for col in train_na.columns[train_na.columns.str.contains('calc')]]
    train_na = train_na.drop(calc_cols,axis=1)
    
    bin_cols = [col for col in train_na.columns[train_na.columns.str.contains('bin')]]
    cat_cols = [col for col in train_na.columns[train_na.columns.str.contains('cat')]]
    ord_cols = [col for col in train_na.columns[train_na.columns.str.contains('ord')]]
    
    bin_cat_train_na = train_na[bin_cols+cat_cols+ord_cols]    
    from fancyimpute import SimpleFill
    bin_cat_train_na_cols = bin_cat_train_na.columns
    bin_cat_train_na = pd.DataFrame(SimpleFill(fill_method='median').complete(bin_cat_train_na.as_matrix()),columns=bin_cat_train_na_cols)

    cont_train_na = train_na.drop(bin_cols+cat_cols+ord_cols,axis=1)
    from fancyimpute import SimpleFill
    cont_train_na_cols = cont_train_na.columns
    cont_train_na = pd.DataFrame(SimpleFill().complete(cont_train_na.as_matrix()),columns=cont_train_na_cols)
    
    train_na_mean_imputed = pd.concat([cont_train_na,bin_cat_train_na],axis=1)
    #train_na_mean_imputed = create_onehot(train_na_mean_imputed)
    return train_na_mean_imputed
```

```python
def pair_above_target(perc,cols):
    worthwhile_pair = []

    for pair in itertools.combinations(cols, 2):
        df = train.groupby(pair)['target'].describe().unstack().unstack()[['mean']]
        if (df > perc).sum().sum() > 1:
            worthwhile_pair.append(pair)

    col_count = {}

    for pair in worthwhile_pair:
        for col in pair:
            if col in col_count:
                col_count[col] += 1
            else:
                col_count[col] = 1
    return col_count
```
